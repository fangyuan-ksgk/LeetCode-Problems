```python
"Grand Counting Solution"
def solveNQueens(n):
    combs = []
    stack = [([],0)]
    while stack:
        cols, layer = stack.pop()
        if layer==n:
            combs.append(cols)
            continue
        poss = [i for i in range(n) if all([i!=cols[r] and abs(i-cols[r])!=abs(layer-r) for r in range(layer)])]    
        for c in poss:
            stack.append((cols+[c],layer+1))
    return len(combs)
```
